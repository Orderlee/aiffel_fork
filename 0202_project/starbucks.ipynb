# Event에 따른 데이터의 양상

```python
df_new = df.drop('became_member_on',axis=1)
```

```python
event_counts_per_id = df_new.groupby('id')['event'].value_counts()

# To display the counts as a DataFrame, you can use unstack()
event_counts_per_id = event_counts_per_id.unstack(fill_value=0)

print(event_counts_per_id)
```

```python
event_counts = event_counts_per_id.sum()

# Plot the counts of each event
plt.bar(event_counts.index, event_counts.values)
plt.xlabel('Events')
plt.ylabel('Count')
plt.title('Count of Each Event')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
ecpd = pd.DataFrame(event_counts_per_id)
```

```python
ecpd
```

```python
relevant_events = event_counts_per_id[['transaction', 'offer completed', 'offer received', 'offer viewed']]

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(relevant_events.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Events and Transactions')
plt.show()
```

offer recieved 하면 offer viewed한 사람과의 연관성이 있는 것 같아보이고, offer completed 한 사람과 transaction도 어느정도 상관이 있어보인다.

```python
ecpd
```

```python
sns.regplot(data=ecpd, x="offer completed", y="transaction")
```
